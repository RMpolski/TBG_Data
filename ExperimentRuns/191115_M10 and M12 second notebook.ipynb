# Temperature sweeps

Notes on the temp sweeps from last time:

Use 10 mA range until 1K. Use 31.6 mA range until 1.4. Then 100 mA till 2K. Keep it there for 2-10K, but you must first turn off the turbo, turn V4 on, V9 off, turn the cernox on, RuOx off, and change the pid control channel to the cernox (channel 5).

Last time, it took 5 hrs to get to 1K (401 gate points, 50 mK increments from 50 mK to 1K (20 steps), 2s wait time between gate points, 180s between temp points). Then it was about 18 minutes per temperature point up to 10K (15 hours for 52 and then 5 hours for 15). There's probably no need to go above 4-5K at 0.1K steps.

Measure 22-20 (M12 with good SC, good fan diagrams) on the lockin865. Measure 26-25 (M10 with good fan diagrams, good correlated insulating states) with lockin830. Use gate range -10 to 10V 401 data points.

First measure with all sensors unplugged for the best measurement.
